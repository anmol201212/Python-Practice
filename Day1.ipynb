```python
# Task 3
# Write an example for different Python data types such as Number(Integer, Float, Complex), String, Boolean, List, Tuple, Set and Dictionary.

#Integer
n = 10
print(n)

# Float
n = 10.5
print(n)

# Complex
n = 1+3j
print(n)

# String 
n = 'anmol chopra'
print(n)

# Boolean 
n = True
print(n)

# List - ordered collection which allows to store different data type items
n = [1,3.5,"anmol", 'anmol chopra']
print(n)

# Tuple - rdered collection of different data types like list but tuples can not be modified once they are created. They are immutable.
n = (1,2,3,4,5)
print(n)

# Set - collection of data types similar to list and tuple. Unlike list and tuple, set is not an ordered collection of items
n = {1,2,3,4,5,6}
print(n)

# Dictionary - unordered collection of data in a key value pair format
n = {1:"Anmol",2:567,3:"Chopra"}
print(n)
```

```python
# Find an Euclidean distance between (2, 3) and (10, 8)
from math import sqrt,dist
a = (2,3)
b = (10,8)

print(sqrt((b[0] - a[0])**2 + (b[1] - a[1])**2))
print(dist(a,b))
```
